```python
#lets try a different approach
#x y got confusing so lets use row v col convention

#lets keep the candidates grid approach because I think it's powerful.

puzzle = [[0, 3, 0, 0, 1, 0, 0, 6, 0],
[7, 5, 0, 0, 3, 0, 0, 4, 8],
[0, 0, 6, 9, 8, 4, 3, 0, 0],
[0, 0, 3, 0, 0, 0, 8, 0, 0],
[9, 1, 2, 0, 0, 0, 6, 7, 4],
[0, 0, 4, 0, 0, 0, 5, 0, 0],
[0, 0, 1, 6, 7, 5, 2, 0, 0],
[6, 8, 0, 0, 9, 0, 0, 1, 5],
[0, 9, 0, 0, 4, 0, 0, 3, 0]]

def view(puz):
    for row in puz:
        print(row)
        
view(puzzle)

```

```python
 def create_cands(puzzle):  
    candidates = [[[],[],[],[],[],[],[],[],[]],
                  [[],[],[],[],[],[],[],[],[]],
                  [[],[],[],[],[],[],[],[],[]],
                  [[],[],[],[],[],[],[],[],[]],
                  [[],[],[],[],[],[],[],[],[]],
                  [[],[],[],[],[],[],[],[],[]],
                  [[],[],[],[],[],[],[],[],[]],
                  [[],[],[],[],[],[],[],[],[]],
                  [[],[],[],[],[],[],[],[],[]]]
    for row in range(0,9):
        for col in range(0,9):
            for k in range(1,10):
                candidates[row][col].append(k)
    for row in range(0,9):
        for col in range(0,9):
            if puzzle[row][col] > 0:
                candidates[row][col] = puzzle [row][col]
    return candidates

            
candidates = create_cands(puzzle)

view(candidates)
```

```python
#there was nothing wrong with our functions to return eligble candidates

def row_grab(row,col,puz=puzzle): #takes the xy coordinates of a sudoku square, returns the nums popd in that row in the puzzle as a list 
    added = []
    for i in puz[row]:
        if int(i) > 0:
            added.append(int(i))
    return(added)

print(row_grab(6,3))

def col_grab(row,col,puz=puzzle): #takes the xy coordinates of a sudoku square, returns the nums popd in that col in the puzzle as a list 
    added = []
    for i in puz:
        if i[col] > 0:
            added.append(i[col])
    return(added)

print(col_grab(6,3))

def xy_trans(row,col): #takes an xy coordinate and gives an xy of the neighbourhood
    return(int((row)/3),int((col)/3))

def neigh_grab(row,col,puz=puzzle): #takes the xy coordinates of a sudoku square, returns the nums popd in that neigh in the puzzle as a list
    nrow,ncol = xy_trans(row,col)
    added = []
    for row in puz[nrow*3:(nrow*3)+3]:
        for i in row[ncol*3:ncol*3+3]:
            if i > 0:
                added.append(i)
    return added

print(neigh_grab(6,3))
print("")
view(puzzle)
```

```python
#they return the correct values. good.

#a function to take a solution and put it into both the puzzle and candidates grid

def cell_solution(row,col,solution,puz=puzzle,cand=candidates):
    puz[row][col] = solution
    cand[row][col] = solution
```

```python
#we can now iterate though the puzzle and find out if any of the cells can be solved

def basic_cell_check(row,col,puz=puzzle,cand=candidates):
    exclude = []
    cands = []
    if puzzle[row][col] > 0:
        return puzzle[row][col]
    exclude.extend(row_grab(row,col,puz))
    exclude.extend(col_grab(row,col,puz))
    exclude.extend(neigh_grab(row,col,puz))
    for val in range(1,10):
        if val not in exclude:
            cands.append(val)
    return cands
    
print(basic_cell_check(0,2))
```

```python
#we can check each puzzle cell for a solution
#if we find one we can update candidates immediately
#we want to loop this check until it finds no solutions on a pass, because solutions found later
#in the loop could inform solutions we missed earlier in the loop

def basic_puzzle_check(puz=puzzle,cands=candidates):
    attempt = 1
    while True:
        sols_found = 0
        for rows in range(0,9):
            for cols in range(0,9):
                if puz[rows][cols] > 0: #dont check solved cells
                    continue
                possible = basic_cell_check(rows,cols,puz,cands)
                if len(possible) == 1:
                    print("Row " + str(rows) + " Col " + str(cols) + " is " + str(possible))
                    cell_solution(rows,cols,possible[0],puz,cands)
                    sols_found += 1
                else:
                    print("Row " + str(rows) + " Col " + str(cols) + " could be " + str(possible))
                    cands[rows][cols] = possible            
        print("Found " + (str(sols_found)) + " solutions on pass no." + (str(attempt)))
        attempt += 1
        if sols_found == 0:
            break
            
basic_puzzle_check(puzzle)
```

```python
#we need to build some candidate analysis function;
#it analyses a set of candidates, and if only one is valid, returns its index
#we can also chuck the basic analysis in here too so if it gets a more complex one it then runs the simple solution
#start with row analysis

def row_analysis(row,puz=puzzle,cands=candidates):
    attempt = 1
    while True:
        anal_set = cands[row]
        solvable = []
        test_set = []
        solved_already = []
        solved_this_attempt = 0
        for entry in anal_set:
            try:
                test_set.extend(entry)
            except:
                test_set.append(entry)
                solved_already.append(entry)
        for test_val in range(1,10):
            if test_set.count(test_val) == 1:
                if test_val not in solved_already:
                    solvable.append(test_val)
                    print(str(test_val) + " is solvable")
                else:
                    print(str(test_val) + " already solved")
        for found_solution in solvable:
            for i in range(0,9):
                if type(anal_set[i]) == list:
                    if found_solution in anal_set[i]:
                        cell_solution(row,i,found_solution,puzzle,candidates)
                        print("Populate " + str(found_solution) + " in column " + str(i))
                        solved_this_attempt += 1
                    else:
                        print("Solution for column " + str(i) + " could be " + str(anal_set[i]))
                else:
                    print("Column " + str(i) + " already solved as " + str(anal_set[i]))
        print("Solved " + str(solved_this_attempt) + " on pass no." + str(attempt))
        attempt += 1
        if solved_this_attempt == 0:
            break
        else:
            basic_puzzle_check(puzzle,candidates)
            
    
    
row_analysis(0,puzzle,candidates)
```
